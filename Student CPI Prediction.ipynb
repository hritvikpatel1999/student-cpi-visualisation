```python
def log_trans(x, inverse = False):     
    if not inverse:         
        return np.log10(x)     
    else:         
        return np.power(10, x) 
 
sns.regplot(df['branch_change'], df['cpi'].apply(log_trans)) 
tick_locs = [10, 20, 50, 100, 200, 500] 
plt.yticks(log_trans(tick_locs), tick_locs);
plt.ylabel('log of cpi');
print('No conclusion from log transformation also')
```
